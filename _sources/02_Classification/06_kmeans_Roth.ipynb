```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
X = make_blobs(n_samples = 400, n_features = 2, centers = 5, cluster_std = 1, random_state=42)
X = X[0]
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
cluster_numbers = [2, 3, 4, 5, 6, 7, 8]
scores = []

fig, ax = plt.subplots(len(cluster_numbers), 1, figsize = (4, 20))
for k,cn in enumerate(cluster_numbers):
    km = KMeans(n_clusters = cn)
    km.fit(X)

    labels = km.labels_
    centroids = km.cluster_centers_
    score = silhouette_score(X, labels)
    scores.append(score)
    
    ax[k].scatter(X[:,0], X[:,1], c = labels, s = 2)
    ax[k].plot(centroids[:,0], centroids[:,1], 'rx')
    ax[k].text(0,-7, f'{score:0.3f}')

plt.show()

```

```python
sil_score = silhouette_score(X, labels)
sil_samples = silhouette_samples(X,labels)

sil_score
```

```python
import requests
from PIL import Image
from io import BytesIO

from matplotlib.colors import ListedColormap

url = "https://raw.githubusercontent.com/YugeTen/Computer-vision-CLAB2/master/peppers.png"

response = requests.get(url)
peppers_image = Image.open(BytesIO(response.content))
peppers = np.array(peppers_image)

print(peppers.shape)
h, w = peppers.shape[:2]

plt.imshow(peppers)
```

```python
x_coords, y_coords = np.meshgrid(range(w), range(h))

x_min, x_max = 190, 200
y_min, y_max = 110, 118

fig, ax = plt.subplots(1,1, figsize = (9,9))
plt.imshow(peppers[y_min:y_max, x_min:x_max,:])

for m, y in enumerate(range(y_min, y_max)):
    for n, x in enumerate(range(x_min, x_max)):
        plt.text(n, m, f'({y},{x})', 
                 fontsize = 6, color ='white', ha = 'center')
plt.show()
```

```python
fig, ax = plt.subplots(1, 3, figsize = (15,5),sharey=True)

zeros = np.zeros(256)
ones = np.ones(256)
color_range = np.linspace(0,1, 256, endpoint=True)

reds =ListedColormap(np.vstack([color_range, zeros, zeros]).transpose())
greens =ListedColormap(np.vstack([zeros, color_range, zeros]).transpose())
blues =ListedColormap(np.vstack([zeros, zeros, color_range]).transpose())

RGB = [reds,greens,blues]

for n in range(3):
    ax[n].imshow(peppers[:,:,n], cmap = RGB[n])
    
plt.show()
```

```python
peppers_dict = dict(
    R = peppers[:,:,0].flatten(),
    G = peppers[:,:,1].flatten(),
    B = peppers[:,:,2].flatten(),
    x = x_coords.flatten(),
    y = y_coords.flatten()   
)


peppers_dict
peppers_df = pd.DataFrame(peppers_dict)
peppers_df['RGB'] = peppers_df.apply(lambda row: [row['R']/255, row['G']/255, row['B']/255], axis=1)
peppers_df.head()
```

First, let's extract the color palette of this image.

```python
X = peppers_df[['R', 'G','B']]

n_clusters = 8
km = KMeans(n_clusters = n_clusters)
km.fit(X)

centroids = km.cluster_centers_
labels = km.labels_

plt.scatter(np.ones(n_clusters), range(n_clusters), c = centroids/255, s = 200, marker = 's')
plt.show()
```

```python
centroids
```

```python
peppers_df['label'] = labels
peppers_df.head()

labels_df = pd.DataFrame({'label':range(n_clusters), 'CR':centroids[:,0]/255, 'CG':centroids[:,1]/255, 'CB':centroids[:,2]/255})
labels_df

peppers_df = peppers_df.merge(labels_df, how = 'left', on = 'label')
```

```python
peppers_RGB = peppers_df[['CR', 'CG', 'CB']].to_numpy()
reshaped_image = peppers_RGB.reshape((384, 512, 3))
plt.imshow(reshaped_image)
plt.axis('off')  # Hide axes for better visualization
plt.show()
```
